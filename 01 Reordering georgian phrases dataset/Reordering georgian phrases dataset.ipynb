```python
# importing pandas, openpyxl and the dataset from a .txt file
import pandas as pd
import openpyxl 
data = pd.read_csv("IntermediateIn.txt", sep=";", engine='python')
data
```

```python
# creating the "Columna" list from de dataset and checking it
columna = data['Columna'].tolist()
```

```python
# creating the "Frase" list from de dataset and checking it
frase = data['Frase'].tolist()
```

```python
# bulding a tuple list from both list: columna and frase
lista = list(zip(columna, frase))
lista
```

```python
# creating empty list for each column I need. 
note = []
cloze = []
sentence = []
pronuntiation = []
translation = []
```

```python
# extracting data from the dataset and assigning it to a specific list
for c,f in lista:
    if c == 'note':
        note.append(f)
    if c == 'cloze':
        cloze.append(f)
    if c == 'sentence':
        sentence.append(f)
    if c == 'pronuntiation':
        pronuntiation.append(f)
    if c == 'translation':
        translation.append(f)
```

```python
# building de dataframe from lists of columns name
df = pd.DataFrame(list(zip(sentence, translation, cloze, pronuntiation, note)), columns = ['sentence', 'translation', 'cloze', 'pronuntiation', 'note'])
df.head()
```

```python
# exporting the dataframe to a .txt file
df.to_excel('IntermediateOut.xlsx', index=False)
```
